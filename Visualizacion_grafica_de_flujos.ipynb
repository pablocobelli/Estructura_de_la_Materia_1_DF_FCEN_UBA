## Como visualizar graficamente el flujo de un fluido

Consideremos como ejemplo de trabajo el flujo bidimensional en el plano asociado a un vortice de circulacion $\Gamma$ constante, ubicado en el origen de coordenadas, dado por:

$$ \vec{u} = \frac{\Gamma}{2\pi r} \: \hat{\theta},$$

en coordenadas cilindricas, o bien mediante:

$$ \vec{u} = \frac{\Gamma}{2\pi\: \big({x^2+y^2}\big)} \big( -y \: \hat{x} + x \: \hat{y}  \big).$$

en coordenadas cartesianas. 

### Definicion del campo de velocidades y la region de visualizacion

Vamos a considerar el flujo en la region del plano cartesiano $x$-$y$ dada por $-1 \leq x, y \leq 1$; para todos nuestros calculos en dicha region usaremos una grilla numerica compuesta de 20$\times$20 puntos. La proxima linea construye la grilla de posiciones `x` e `y`:

```python
x, y = np.meshgrid(np.linspace(-1, 1, 20), np.linspace(-1, 1, 20)) 
```

Ahora damos un valor a la constante $\Gamma$, en este caso la hacemos igual a 1.

```python
Gamma = 1
```

Y finalmente definimos el campo vectorial en coordenadas cartesianas, definiendo por separado cada una de sus dos componentes, que llamamos `ux` y `uy`. Para ello, hacemos:

```python
ux = -Gamma/2/np.pi/(x**2+y**2)*y
uy = +Gamma/2/np.pi/(x**2+y**2)*x
```

### Como graficar un campo vectorial

El comando [`quiver`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.quiver) del modulo `matplotlib` de Python permite graficar un campo vectorial 2D de componentes `(ux, uy)` en una grilla cartesiana del espacio `(x,y)`. La segunda linea del codigo que sigue da un ejemplo de su uso en el caso del campo que estamos estudiando. El resto de las lineas crean la figura, le agregan titulos a los ejes, una grilla para referencia y se aseguran de que la relacion de aspecto entre ambos ejes sea la misma.

```python
# Graficamos.
plt.figure(figsize=(5,5))
plt.quiver(x, y, ux, uy, color='r')
plt.plot(0,0,'ko')
plt.xlabel('x')
plt.ylabel('y')
plt.grid('on')
plt.axes().set_aspect('equal')
```

### Como graficar lineas de corriente

El comando [`streamplot`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.streamplot) del modulo `matplotlib` de Python permite graficar algunas lineas de corriente asociadas a un campo vectorial 2D de componentes `(ux, uy)` en una grilla cartesiana del espacio dada por `(x,y)`. La segunda linea del codigo que sigue da un ejemplo de su uso en el caso del campo que estamos estudiando. El resto de las lineas crean la figura, le agregan titulos a los ejes, una grilla para referencia y se aseguran de que la relacion de aspecto entre ambos ejes sea la misma.

```python
# Graficamos.
plt.figure(figsize=(5,5))
plt.streamplot(x, y, ux, uy, density=1)
plt.plot(0,0,'ko')
plt.xlabel('x')
plt.ylabel('y')
plt.grid('on')
plt.axes().set_aspect('equal')
```

# Otros flujos interesantes

## Fuente lineal de caudal constante

$$ \vec{u} = \frac{Q}{2\pi r} \: \hat{r},$$

en coordenadas cilindricas, o bien mediante:

$$ \vec{u} = \frac{Q}{2\pi \sqrt{x^2+y^2}} \big( x \: \hat{x} + y \: \hat{y}  \big).$$

en coordenadas cartesianas. 

```python
# Definimos los parametros del campo.
Q     = 1

# Construimos la grilla sobre la que graficaremos.
x, y = np.meshgrid(np.linspace(-1, 1, 20), np.linspace(-1, 1, 20)) 

# Definimos el campo asociado a la fuente.
ux = +Q/2/np.pi/np.sqrt(x**2+y**2)*x
uy = +Q/2/np.pi/np.sqrt(x**2+y**2)*y

# Graficamos conjuntamente el campo y una linea de corriente.
fig = plt.figure(figsize=(5,5))
plt.quiver(x, y, ux, uy, color='r')
plt.streamplot(x, y, ux, uy, density=1.25, color='b')
plt.plot(0,0,'ko')
plt.xlabel('x')
plt.ylabel('y')
plt.grid('on')
plt.xlim(-1,1);
plt.ylim(-1,1);
plt.axes().set_aspect('equal')
```

## Fuente lineal de caudal constante y vortice lineal de circulacion constante

Ahora consideremos el caso de una fuente lineal de caudal constante $Q$ superpuesta a un vortice lineal de circulacion
constante $\Gamma$, ambos ubicados en el mismo punto, que por practicidad consideraremos el origen de nuestro sistema de coordenadas. Dicho flujo se describe mediante el campo de velocidades dado por:

$$\vec{u} = \frac{Q}{2\pi r} \: \hat{r} + \frac{\Gamma}{2\pi r} \: \hat{\theta}$$

en coordenadas cilindricas. En coordenadas cartesianas, mas adaptadas para graficar en una grilla como la que tenemos, podemos reescribir el campo de velocidades como

$$ \vec{u} = \frac{Q}{2\pi (x^2+y^2)} \: (x \: \hat{x} + y \: \hat{y}) + \frac{\Gamma}{2\pi (x^2+y^2)} (-y \: \hat{x} + x \: \hat{y}),$$

y finalmente, reagrupando terminos, obtenemos la expresion mas concisa dada por:

$$ \vec{u} = \frac{1}{2\pi (x^2+y^2)} \: \bigg[ \big(Q\: x- \Gamma\: y\big) \: \hat{x} + \big(Q \:y + \Gamma\: x \big) \: \hat{y} \bigg] $$

Las lineas que siguen definen los parametros del campo (valores concretos de $Q$ y $\Gamma$), construyen la grilla cartesiana para el calculo, definen el campo total como la suma de los campos asociados a la fuente y al vortice, y finalmente contruyen un grafico en donde se superponen el campo vectorial y las lineas de corriente.

```python
# Definimos los parametros del campo.
Q     = 0.2
Gamma = 1

# Construimos la grilla sobre la que graficaremos.
x, y = np.meshgrid(np.linspace(-1, 1, 20), np.linspace(-1, 1, 20)) 

# Definimos el campo asociado al vortice.
ux_vort = -Gamma/2/np.pi/np.sqrt(x**2+y**2)*y
uy_vort = +Gamma/2/np.pi/np.sqrt(x**2+y**2)*x

# Definimos el campo asociado a la fuente.
ux_fuen = +Q/2/np.pi/np.sqrt(x**2+y**2)*x
uy_fuen = +Q/2/np.pi/np.sqrt(x**2+y**2)*y

# El campo total es la suma de ambos.
ux_tot  = ux_vort + ux_fuen
uy_tot  = uy_vort + uy_fuen

# Graficamos conjuntamente el campo y una linea de corriente.
fig = plt.figure(figsize=(5,5))
plt.quiver(x, y, ux_vort + ux_fuen, uy_vort + uy_fuen, color='r')
plt.streamplot(x, y, ux_vort + ux_fuen, uy_vort + uy_fuen, density=1.25, color='b')
plt.plot(0,0,'ko')
plt.xlabel('x')
plt.ylabel('y')
plt.grid('on')
plt.xlim(-1,1);
plt.ylim(-1,1);
plt.axes().set_aspect('equal')
```

## Dos fuentes lineales separadas por una distancia d

Consideremos ahora las lineas de flujo asociadas a dos fuentes (o sumideros) lineales separadas por una distancia $d$.

En coordenadas cartesianas, el campo de velocidades plano asociado a este flujo se expresa, por ejemplo, como:

$$\vec{u} = \frac{Q_1}{2\pi} \frac{1}{(x-d/2)^2 + y^2} \bigg[ (x-d/2) \: \hat{x} + y \: \hat{y} \bigg] + \frac{Q_2}{2\pi} \frac{1}{(x+d/2)^2+y^2} \bigg[ (x+d/2) \: \hat{x} + y \: \hat{y} \bigg].$$

Definimos los parametros del campo, es decir, los valores de $Q_1$, $Q_2$ y $d$. Luego construimos la grilla espacial sobre la que vamos a calcular los valores del campo. Finalmente definimos las componentes del campo; en este caso lo hacemos definiendo el campo de cada fuente por separado. Esto ultimo lo logramos con las siguientes lineas de codigo:

```python
# Definimos los parametros del campo.
Q1 = +1
Q2 = +1
d  = 5

# Construimos la grilla sobre la que graficaremos.
x, y = np.meshgrid(np.linspace(-10, 10, 30), np.linspace(-10, 10, 30)) 

# Definimos el campo asociado a la fuente en (d/2,0).
u1x = Q1/2/np.pi/((x-d/2)**2+y**2)*(x-d/2)
u1y = Q1/2/np.pi/((x-d/2)**2+y**2)*y

# Definimos el campo asociado a la fuente en (-d/2,0).
u2x = Q2/2/np.pi/((x+d/2)**2+y**2)*(x+d/2)
u2y = Q2/2/np.pi/((x+d/2)**2+y**2)*y
```

Ahora vamos a graficar conjuntamente el campo vectorial y las lineas de corriente para dos fuentes lineales, y para una fuente y un sumidero; considerando en ambos casos los mismos caudales para cada fuente y/o sumidero. Las siguientes lineas construyen los 4 graficos que se muestran mas abajo. 

```python
# Graficamos 4 subfiguras en una sola figura.
fig = plt.figure(figsize=(8,8))

# Graficamos el campo vectorial para dos fuentes (de igual caudal).
ax1 = fig.add_subplot(221)
ax1.quiver(x, y, u1x + u2x, u1y + u2y, color='r')
ax1.plot(-d/2,0,'ko')
ax1.plot(+d/2,0,'ko')
plt.xlabel('x')
plt.ylabel('y')
plt.grid('on')
plt.title('DOS FUENTES');

# Graficamos el campo vectorial para una fuente y un sumidero (de igual caudal).
ax2 = fig.add_subplot(222)
ax2.quiver(x, y, u1x - u2x, u1y - u2y, color='r')
ax2.plot(-d/2,0,'ko')
ax2.plot(+d/2,0,'ko')
plt.xlabel('x')
plt.ylabel('y')
plt.grid('on')
plt.xlim(-10,10);
plt.ylim(-10,10);
plt.title('UNA FUENTE Y UN SUMIDERO');

# Graficamos las lineas de corriente para dos fuentes (de igual caudal).
ax3 = fig.add_subplot(223)
ax3.streamplot(x, y, u1x + u2x, u1y + u2y, color='b', density=1.5)
ax3.plot(-d/2,0,'ko')
ax3.plot(+d/2,0,'ko')
plt.xlabel('x')
plt.ylabel('y')
plt.grid('on')
plt.xlim(-10,10);
plt.ylim(-10,10);

# Graficamos las lineas de corriente para una fuente y un sumidero (de igual caudal).
ax4 = fig.add_subplot(224)
ax4.streamplot(x, y, u1x - u2x, u1y - u2y, color='b', density=1.5)
ax4.plot(-d/2,0,'ko')
ax4.plot(+d/2,0,'ko')
plt.xlabel('x')
plt.ylabel('y')
plt.grid('on')
plt.xlim(-10,10);
plt.ylim(-10,10);
```

## Fuente lineal de caudal constante frente a corriente uniforme

Consideremos ahora el flujo bidimensional en el plano asociado a una fuente lineal de caudal $Q$ constante, ubicado en el origen de coordenadas, que enfrenta a una corriente uniforme de velocidad $u_0$ en la direccion $\hat{x}$. El campo de velocidades asociado a este flujo puede escribirse como:

$$ \vec{u} = \frac{Q}{2\pi r} \: \hat{r} + u_0 \: \hat{x}$$

en coordenadas mixtas, o bien

$$ \vec{u} = \frac{Q}{2\pi\: \big({x^2+y^2}\big)} \big[(u_0 -y) \: \hat{x} + x \: \hat{y}  \big]$$

en coordenadas cartesianas.

Mediante las siguientes lineas de codigo definimos los valores numericos para los parametros del campo (en este caso, $u_0$ y $Q$), construimos la grilla numerica que emplearemos para calcular y definimos el campo de velocidades como la suma de los campos de la fuente y de la corriente uniforme. 

```python
# Definimos los parametros del campo.
u0  = 1
Q   = 10

# Construimos la grilla sobre la que graficaremos.
x, y = np.meshgrid(np.linspace(-10, 10, 25), np.linspace(-10, 10, 25)) 

# Definimos el campo asociado a la fuente en (0,0).
u1x = Q/2/np.pi/(x**2+y**2)*x
u1y = Q/2/np.pi/(x**2+y**2)*y

# Definimos el campo asociado a la corriente uniforme de velocidad u0 en x.
u2x = u0
u2y = 0

# El campo total es la suma de ambos.
ux_tot = u1x + u2x
uy_tot = u1y + u2y
```

Finalmente, graficamos conjuntamente el campo vectorial (en flechas rojas) y las lineas de corriente (en azul) para este flujo. 

```python
# Graficamos.
fig = plt.figure(figsize=(5,5))
plt.quiver(x, y, ux_tot, uy_tot, color='r')
plt.streamplot(x, y, ux_tot, uy_tot, color='b', density=1.5)
plt.plot(0,0,'ko')
plt.xlabel('x')
plt.ylabel('y')
plt.grid('on')
plt.xlim(-10,10);
plt.ylim(-10,10);
plt.axes().set_aspect('equal')
plt.title('FUENTE Y CORRIENTE UNIFORME');
```

## Fuente y sumidero lineales frente a corriente uniforme

Como ultimo ejemplo interesatnte, exploremos el flujo bidimensional en el plano asociado a un par fuente--sumidero de caudales $\pm Q$ constantes, separados sus componentes por una distancia $2\:d$, y que enfrenta a una corriente uniforme de velocidad $u_0$ en la direccion $\hat{x}$. El campo de velocidades asociado a este flujo se define mediante la expresion

$$ \vec{u} = \bigg\{ \frac{Q}{2\pi } \: \bigg[ \frac{x+d}{(x+d)^2+y^2} - \frac{x-d}{(x-d)^2+y^2}  \bigg] + u_0 \: \hat{x}        \bigg\} + \frac{Q}{2\pi } \: \bigg[ \frac{y}{(x+d)^2+y^2} - \frac{y}{(x-d)^2+y^2}  \bigg] \: \hat{y},$$

en coordenadas cartesianas.

```python
# Definimos los parametros del campo.
Q   = 15
d   = 5/2
u0  = 1

# Construimos la grilla sobre la que graficaremos.
x, y = np.meshgrid(np.linspace(-10, 10, 30), np.linspace(-10, 10, 30)) 

# Definimos el campo asociado al sumidero en (d,0).
u1x = -Q/2/np.pi/((x-d)**2+y**2)*(x-d)
u1y = -Q/2/np.pi/((x-d)**2+y**2)*y

# Definimos el campo asociado a la fuente en (-d,0).
u2x = Q/2/np.pi/((x+d)**2+y**2)*(x+d)
u2y = Q/2/np.pi/((x+d)**2+y**2)*y

# Definimos el campo asociado a la corriente uniforme de velocidad u0 en x.
u3x = u0
u3y = 0
```

```python
# Graficamos.
fig = plt.figure(figsize=(5,5))
# ax1 = fig.add_subplot(111)
plt.quiver(x, y, u1x + u2x + u3x, u1y + u2y +u3y, color='r')
plt.streamplot(x, y, u1x + u2x + u3x, u1y + u2y + u3y, color='b', density=1.5)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(-d/2,0,'ko')
plt.plot(+d/2,0,'ko')
plt.grid('on')
plt.xlim(-10,10);
plt.ylim(-10,10);
plt.axes().set_aspect('equal')
plt.title('FUENTE Y SUMIDERO FRENTE A CORRIENTE UNIFORME');
```
