## Problema

La temperatura en un tunel viene dada por la expresion:

$$ T(x,t) = T_0 - \alpha \: \text{e}^{-x/L} \: \sin \left(\frac{2\pi}{\tau} t \right) $$

donde $T_0, \alpha, L$ y $\tau$ son constantes positivas. Una particula de fluido se mueve en el interior del tunel con velocidad constante $U$. 

En estas condiciones:

(a) Hallar la variación de temperatura por unidad de tiempo que experimenta la partícula bajo una descripción euleriana. Graficar la temperatura para instantes próximos e interpretar las componentes de la derivada total.

(b) Idem que (a), pero desde una descripción lagrangiana. Coinciden las dos descripciones realizadas?

----

## Resolucion analitica

### Observaciones iniciales acerca del campo de temperaturas en el tunel

Comencemos por notar que la funcion $T$ da, para todo instante de tiempo $t$, la temperatura $T$ en cualquier punto $x$ del tunel; es decir, se trata de una descripcion **de campo** de la temperatura.

Ahora observemos la forma funcional de $T(x,t)$. 
- Si consideramos un lugar fijo del tunel, la temperatura local parece oscilar armonicamente en torno a $T_0$ con una frecuencia dada por $f=1/\tau$. 
- Por otro lado, si consideramos un instante fijo en el tiempo, vemos que el perfil de temperaturas a lo largo del tunel tiende exponencialmente hacia un valor asintotico $T_0$. 

A partir de estas consideraciones, podemos intentar responder en forma cualitativamente correcta a la siguiente pregunta: **para tiempos suficientemente largos ($t\rightarrow+\infty$), que comportamiento presenta la variacion temporal de la temperatura de una particula fluida?** [Piensen la respuesta antes de continuar leyendo!]

La respuesta es sencilla: a tiempos largos, la particula se habra alejado tanto del origen de coordenadas que su temperatura variara muy poco respecto de $T_0$. Es decir, **para tiempos suficientemente largos esperamos que la variacion de la temperatura siguiendo a la particula sea nula.**

### Empleando la descripcion lagrangiana

Como dijimos inicialmente, la expresion que tenemos para la temperatura $T(x,t)$ es una expresion **de campo**, es decir, euleriana. Lo que buscamos, sin embargo, es obtener la variacion temporal de la temperatura de **una particula dada**, para lo cual resulta necesario disponer de una descripcion **lagrangiana**. 

En la descripcion lagrangiana, la posicion instantanea $X$ de una particula que a tiempo $t_0$ se encontraba en $x_0$ se denota mediante

$$X(t \: ; x_0),$$

que resulta una funcion de *una sola variable* (el tiempo $t$) y de un parametro, $x_0$, que sirve para identificar univocamente a la particula que estamos considerando. Podemos hacer esto mas evidente empleando la notacion equivalente:

$$X(t \: ; x_0) \equiv X_{x_0}(t).$$

En este caso conocemos *a priori* la trayectoria de la particula, dado que nos dicen que se mueve en el tunel con velocidad constante $U$. Asumiendo que partio de $x_0$ a $t=0$, podemos escribir entonces:

$$X(t; x_0) = Ut + x_0.$$

Luego calcular la temperatura instantanea de la particula de fluido en su movimiento en el tunel resulta sencillo: basta con reemplazar esta trayectoria en la coordenada espacial del campo $T(x,t)$. Haciendo esto se obtiene:

$$ T(t;x_0) = T_0 - \alpha \: \text{e}^{-(Ut+x_0)/L} \: \sin \left(\frac{2\pi}{\tau} t \right), $$

lo que nos da la temperatura **en la descripcion lagrangiana**, es decir, siguiendo a una particula dada (en este caso, aquella que a $t=0$ se encontraba en $x_0$).  

Lo que resta es calcular la variacion temporal de la temperatura de la particula, que surge naturalmente de considerar:

$$\frac{d}{dt}\big[T(t; x_0)\big] \equiv \dot{T}(t; x_0).$$

Notemos que se trata de una derivada **total** puesto que la temperatura *lagrangiana* corresponde a una particula material, y solo experimenta variaciones *temporales* conforme se mueve en el tunel. 

La variacion temporal de la temperatura de la particula resulta finalmente:

$$\begin{align}
\dot{T}(t; x_0) &= \alpha \: \frac{U}{L} \: \text{e}^{-(Ut+x_0)/L} \: \sin 
\bigg( \frac{2\pi}{\tau} t\bigg) - \alpha \: \text{e}^{-(Ut+x_0)/L} \: \cos \bigg( \frac{2\pi}{\tau} t\bigg) \: \frac{2\pi}{\tau}  \\
 &= \alpha \: \text{e}^{-(Ut+x_0)/L} \: \bigg[  \frac{U}{L} \: \sin 
\bigg( \frac{2\pi}{\tau} t\bigg) - \frac{2\pi}{\tau} \: \cos \bigg( \frac{2\pi}{\tau} t\bigg) \bigg].
\end{align}$$

*[Notemos que las unidades son correctas: $\alpha$ tiene unidades de temperatura, y $U/L$ tiene unidades de inversa de tiempo, asi como $1/\tau$.]*

Vemos entonces que **la variacion temporal de la temperatura que experimenta la particula tiende a cero a medida que la particula se mueve en el tunel**, es decir, para tiempos suficientemente largos ($t \rightarrow +\infty$).  

## Empleando la descripcion euleriana

Dado que partimos de una descripcion de campo para la temperatura $T(x,t)$,  la variacion (temporal) de la temperatura **siguiendo a una particula material** se obtiene de forma inmediata empleando el **operador derivada material**, definido como:

$$\frac{D}{Dt} \equiv \frac{d}{dt} = \frac{\partial}{\partial t} + \left( \vec{u}\cdot\vec{\nabla} \right),$$
o, en notacion indicial:
$$\frac{D}{Dt} \equiv \frac{d}{dt} = \frac{\partial}{\partial t} +{u}_j \partial_j .$$

En nuestro caso, para la temperatura tenemos:

$$\frac{dT}{dt}(x,t) = \frac{\partial}{\partial t}T(x,t) + \left( \vec{u}\cdot\vec{\nabla} \right) T(x,t)$$

Calculemos cada uno de los sumandos a la derecha de esta ultima expresion. El primero de ellos resulta igual a 

$$\frac{\partial}{\partial t}T(x,t) = - \: \alpha \: \text{e}^{-x/L} \: \cos \bigg( \frac{2\pi}{\tau} t \bigg) \: \frac{2\pi}{\tau}$$

Este termino que acabamos de calcular da cuenta de la variacion temporal de la temperatura de la particula **debida exclusivamente al hecho de que el campo de temperatura esta variando en el tiempo**. 

Calculemos ahora el segundo termino. Para ello consideramos

$$\begin{align}
\left( \vec{u}\cdot\vec{\nabla} \right) T(x,t) &= U \: \frac{\partial}{\partial x} T(x,t) \\
&= U \: \frac{\alpha}{L} \: \text{e}^{-x/L} \: \sin \bigg( \frac{2\pi}{\tau} t\bigg)   
\end{align}$$

Este termino **de caracter puramente espacial** y que contiene a la velocidad del flujo describe el **transporte del campo de temperaturas por el flujo**, y recibe el nombre de *termino convectivo*. 

Finalmente, la derivada material completa resulta igual a:

$$\begin{align}
\frac{dT}{dt}(x,t) &= - \: \alpha \: \text{e}^{-x/L} \: \cos \bigg( \frac{2\pi}{\tau} t \bigg) \: \frac{2\pi}{\tau} + U \: \frac{\alpha}{L} \: \text{e}^{-x/L} \: \sin \bigg( \frac{2\pi}{\tau} t\bigg) \\
&= \alpha \: \text{e}^{-x/L} \: \bigg[  \frac{U}{L} \: \sin 
\bigg( \frac{2\pi}{\tau} t\bigg) - \frac{2\pi}{\tau} \: \cos \bigg( \frac{2\pi}{\tau} t\bigg) \bigg].
\end{align}$$

Ahora bien, para que esta expresion corresponda a la variacion temporal **siguiendo a una particula**, hace falta especificar de que particula estamos hablando. Para ello, basta con especificar la trayectoria de la particula considerada, y reemplazar esta ultima en la variable espacial $x$. 

En nuestro caso, vamos a continuar considerando la misma particula que tomamos en la seccion anterior, es decir, aquella que se mueve con velocidad $U$ y que a $t=0$ se encontraba en $x_0$. Es decir:

$$X_{x_0}(t) = X(t; x_0) = Ut+x_0.$$

Entonces haciendo $\: x = X(t;x_0 ) =Ut+x_0 \:$ en la ultima expresion, obtenemos:

$$\frac{dT}{dt}(X(t;x_0),\:t) = \dot{T}(t; x_0) = \alpha \: \text{e}^{-(Ut+x_0)/L} \: \bigg[  \frac{U}{L} \: \sin 
\bigg( \frac{2\pi}{\tau} t\bigg) - \frac{2\pi}{\tau} \: \cos \bigg( \frac{2\pi}{\tau} t\bigg) \bigg].$$


Observemos que esta ultima expresion, obtenida desde una descripcion euleriana, coincide exactamente con la calculada a traves de la descripcion lagrangiana en la seccion precedente.

## Ilustracion numerica

En esta seccion vamos a ilustrar numericamente los resultados obtenidos. Para ello, comencemos por cargar los modulos de Python que necesitaremos tanto para el calculo numerico como para la generacion de graficos. (Para mas informacion, ver la ultima seccion de este documento).

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Para comenzar, vamos a elegir valores concretos para los parametros del problema. No nos vamos a preocupar acerca de las unidades *concretas* de estos parametros, sino que vamos a asumir que las mismas estan definidas en el marco de un mismo sistema de unidades consistente.

Elegimos entonces, por ejemplo:

```python
To    = 2
alpha = 1
L     = 0.5
tau   = 0.5
U     = 2
xo    = 0
```

Dado que se trata de un calculo numerico (no analitico), debemos especificar en que puntos espaciales e instantes temporales **concretos** estaremos evaluando nuestras expresiones, en particular el campo de temperaturas $T(x,t)$. Para ello, definimos las grillas espaciales y temporales:

```python
# Construimos las grillas espaciales y temporales, 
#   cada una de ellas compuesta de 100 puntos (este valor es arbitrario).
Npx   = 100
Npt   = 100
xv    = np.linspace(0, 2, Npx)
tv    = np.linspace(0, 1, Npt)
x, t  = np.meshgrid(xv, tv)
```

Construimos entonces el campo de temperaturas de acuerdo a la expresion del enunciado. 

```python
T = To - alpha*np.exp(-x/L)*np.sin(2*np.pi*t/tau)
```

Ahora vamos a graficar el campo de temperaturas obtenido para esta eleccion particular de los valores de los parametros `To, alpha, L` y `tau`.

```python
# Graficamos.
plt.pcolor(x, t, T)
plt.title(r'$T(x, t)$', fontsize=18);
plt.xlabel(r'$x$', fontsize=18);
plt.ylabel(r'$t$', fontsize=18);
plt.colorbar()
plt.show()
```

Este grafico que acabamos de construir se denomina habitualmente **diagrama espaciotemporal**. Nos permite representar sinteticamente los valores de una cantidad escalar (la temperatura $T$) codificados en color para cada punto $x$ y tiempo $t$ simultaneamente. Observen que el tiempo avanza hacia arriba y la coordenada espacial aumenta hacia la derecha. La temperatura, representada en color, toma valores entre 1 (azul) y 3 (rojo), en torno del valor central 2 (verde). 

Si 'nos paramos' en un punto fijo del tunel, observariamos unicamente la variacion temporal del campo de temperaturas. Esto corresponde a hacer un corte vertical sobre el diagrama espaciotemporal, y graficar los valores de temperatura que yacen sobre ese corte en funcion del tiempo. Hagamos esto ultimo para distintos puntos fijos $x_*$ en el tunel:

```python
# Graficamos.
fig = plt.figure(figsize=(13,3.5))

ax1 = fig.add_subplot(121)
Tmap = ax1.pcolor(x, t, T)
plt.colorbar(Tmap,ax=ax1) 
ax1.plot(0.25*np.ones((len(tv),1)), tv,'k--')
ax1.plot(0.5*np.ones((len(tv),1)), tv,'k:')
ax1.plot(1*np.ones((len(tv),1)), tv,'k-.')
plt.title(r'$T(x, t)$', fontsize=18);
plt.xlabel(r'$x$', fontsize=18);
plt.ylabel(r'$t$', fontsize=18);

ax2 = fig.add_subplot(122)
ax2.plot(tv, T[:,24],'k--', label=r'$x_* = 0.25$')
ax2.plot(tv, T[:,49],'k:', label=r'$x_* = 0.50$')
ax2.plot(tv, T[:,99],'k-.', label=r'$x_* = 1.00$')
plt.ylim(1,3);
plt.grid('on')
plt.xlabel(r'$t$', fontsize=18);
plt.ylabel(r'$T(x_*, t)$', fontsize=18);
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=16);
```

En los graficos sobre estas lineas vemos este comportamiento. Si permanecemos quietos en una posicion fija del tunel,  
observamos un cambio **temporal** en la temperatura local. Las lineas verticales en el grafico de la izquierda se corresponden
a las lineas en el grafico de la derecha, *respetando los simbolos empleados en cada caso*. Para posiciones fijas sobre el tunel, solo vemos la variacion temporal del campo, cuya amplitud decrece conforme nos alejamos del origen de coordenadas.

Si ahora 'nos paramos' en un instante fijo, observariamos unicamente la variacion espacial del campo de temperaturas. Esto corresponde a hacer un corte horizontal sobre el diagrama espaciotemporal, y graficar los valores de temperatura que yacen sobre ese corte en funcion del tiempo. Hagamos esto ultimo para distintos instantes $t_*$:

```python
# Graficamos.
fig = plt.figure(figsize=(13,3.5))

ax1 = fig.add_subplot(121)
Tmap = ax1.pcolor(x, t, T)
plt.colorbar(Tmap,ax=ax1) 
ax1.plot(xv, 0.1*np.ones((len(xv),1)), 'k--')
ax1.plot(xv, 0.5*np.ones((len(xv),1)), 'k:')
ax1.plot(xv, 0.9*np.ones((len(xv),1)),'k-.')
plt.title(r'$T(x, t)$', fontsize=18);
plt.xlabel(r'$x$', fontsize=18);
plt.ylabel(r'$t$', fontsize=18);

ax2 = fig.add_subplot(122)
ax2.plot(xv, T[9,:],'k--', label=r'$t_* = 0.1$')
ax2.plot(xv, T[49,:],'k:', label=r'$t_* = 0.5$')
ax2.plot(xv, T[89,:],'k-.', label=r'$t_* = 0.9$')
plt.ylim(1,3);
plt.grid('on')
plt.xlabel(r'$t$', fontsize=18);
plt.ylabel(r'$T(x, t_*)$', fontsize=18);
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=16);
```

Los graficos sobre estas lineas muestran este comportamiento. Si observamos la distribucion de temperatura en el tunel en diferentes instantes fijos, vemos el cambio **espacial** en la temperatura instante a instante. Las lineas horizontales en el grafico de la izquierda se corresponden
a las lineas en el grafico de la derecha, *respetando los simbolos empleados en cada caso*. Para instantes fijos, solo vemos la variacion espacial del campo, cuya amplitud decrece conforme avanza el tiempo.

### A partir de esto, como podemos visualizar entonces lo que sucede con una particula material? 

Si bien los dos casos considerados anteriormente nos ayudan a comprender la evolucion del campo de temperaturas, ninguno de ellos (variacion temporal a posicion fija, o variacion espacial a tiempo fijo) corresponden a lo que sucede con las particulas materiales que se mueven en el tunel. 

La particula material que estamos **siguiendo** no permanece fija ni en el tiempo ni en el espacio. En particular, sabemos que la particula sigue una trayectoria dada por $X(t;x_0) = Ut+x_0$. En el diagrama espaciotemporal de $T(x,t)$, esta trayectoria se corresponde con una recta oblicua (i.e., ni horizontal ni vertical) en el diagrama. Si graficamos la temperatura sobre dicha recta, tendremos entonces **la temperatura de la particula siguiendola en su movimiento**:  

```python
# Definimos la trayectoria de la particula.
xp = U*tv+xo
# Definimos el perfil de temperatura sobre dicha trayectoria.
Tp = To - alpha*np.exp(-xp/L)*np.sin(2*np.pi*tv/tau)
```

```python
# Graficamos.
fig = plt.figure(figsize=(13,3.5))

ax1 = fig.add_subplot(121)
Tmap = ax1.pcolor(x, t, T)
plt.colorbar(Tmap,ax=ax1) 
ax1.plot(U*t+xo,t,'w:')
plt.title(r'$T(x, t)$', fontsize=18);
plt.xlabel(r'$x$', fontsize=18);
plt.ylabel(r'$t$', fontsize=18);

ax2 = fig.add_subplot(122)
ax2.plot(tv, Tp,'k--', label=r'Particula con $X(t;x_0)$')
plt.ylim(1,3);
plt.grid('on')
plt.xlabel(r'$t$', fontsize=18);
plt.ylabel(r'$T(X(t;x_0), t)$', fontsize=18);
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=16);
```

Siguiendo entonces la trayectoria de la particula (la linea blanca en el diagrama espaciotemporal), vemos que
la temperatura de la particula tiende a $T_0$ para tiempos suficientemente largos. Esto significa entonces que la
**variacion** de la temperatura **siguiendo a la particula** tiende a cero en ese mismo limite, resultado que 
ya habiamos obtenido analiticamente en las secciones precedentes.
