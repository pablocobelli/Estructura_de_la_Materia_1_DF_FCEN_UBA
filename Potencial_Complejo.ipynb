## Introducción

$$\newcommand{\ver}[1]{{\hat{#1}}}
\renewcommand{\div}{\vec{\nabla}\cdot}
\newcommand{\rot}{\vec{\nabla}\times}
\newcommand{\grad}{\vec{\nabla}}
\newcommand{\td}[2]{\frac{d #1}{d #2}}
\newcommand{\pd}[2]{\frac{\partial #1}{\partial #2}}
\newcommand{\lap}{\nabla^{2}}
\renewcommand{\u}{\vec{u}}
\renewcommand{\v}{\vec{v}}$$

En el marco de lo que vieron en las clases teóricas, las condiciones de flujo plano, incompresible e irrotacional de los flujos nos permiten introducir dos funciones o campos escalares, $\phi$ y $\psi$:

\begin{align}
\div \u &= 0 \quad \Rightarrow \quad \u = \grad \phi, \\
\rot \u &= 0 \quad \Rightarrow \quad \u = \grad \psi \times \ver{k}.
\end{align}

Los campos $\phi(\vec{x})$ y $\psi(\vec{x})$ se denominan *potencial de velocidades* y 
*función corriente*, respectivamente. En la segunda de estas ecuaciones, hemos elegido trabajar en el plano $x$-$y$, y el símbolo $\ver{k}$ representa entonces el versor en la dirección perpendicular a dicho plano. 

Asimismo, es fácil ver que estas funciones satisfacen las condiciones de Cauchy-Riemann, 

\begin{align}
u_{x} = \pd{\phi}{x} &= + \pd{\psi}{y},  \\
u_{y} = \pd{\phi}{y} &= -\pd{\psi}{x}, 
\end{align}

y resultan armónicas, i.e., $\lap \phi = \lap \psi = 0$ (para ello, basta con reexpresar la incompresibilidad e irrotacionalidad del flujo en términos de $\phi$ y $\psi$). 

## Aclaración acerca de los flujos singulares

A lo largo de esta guía de ejercicios, estaremos considerando algunos flujos que presentan singularidades que corresponden, p.ej., a la presencia de una fuente puntual de caudal constante (como el caso que vamos a analizar aquí) o de un vórtice de circulación dada, o de combinaciones de ellos. En todos estos casos, está sobreentendido que las condiciones establecidas por las ecuaciones precedentes sobre la divergencia y el rotor del campo de velocidades son válidas **únicamente en todo punto del plano en que no haya sido introducida una singularidad**. 

## Ilustración de la determinación del potencial complejo

Vamos a considerar el caso del flujo producido por una singularidad puntual que emite isótropamente un caudal $Q$ (por unidad de longitud perpendicular al plano). Adoptaremos, por comodidad, un sistema de referencia centrado en la singularidad, y describiremos el flujo en el plano empleando coordenadas cilíndricas. La descripción obtenida será válida en todo punto del plano excepto en la posición de la singularidad.

A partir de la ecuación para $\psi$, y empleando la expresión para el gradiente en coordenadas cilíndricas, obtenemos

\begin{align}
u_{r} &= \frac{1}{r} \pd{\psi}{\theta}, \\
u_{\theta} &= - \pd{\psi}{r}.
\end{align}   

Dado que el flujo es isótropo, la magnitud de la velocidad no puede depender de la coodenada angular. Asimismo, la condición de incompresibilidad establece que el flujo debe tener caudal neto igual a $Q$ a través de toda curva cerrada que rodee la singularidad. Esto último implica que la componente radial debe ser distinta de cero.
Si consideramos entonces el flujo a través de una curva circular cerrada de radio $r$ centrada en el origen, obtenemos la condición:

\begin{equation*}
\oint \: (u_{r} \ver{r}) \cdot d\vec{s} =  
\oint \: (u_{r} \ver{r}) \cdot \ver{r} \: d\theta = 
\int_{0}^{2\pi} u_{r} \: r \: d\theta = 2\pi u_{r} r = Q,
\end{equation*}

luego la componente radial del campo de velocidades resulta, como esperábamos, 

\begin{equation}
u_{r} = \frac{Q}{2\pi r}.
\end{equation}

Con el resultado provisto por la última igualdad, que ahora resulta

\begin{equation}
u_{r} = \frac{Q}{2\pi r} = \frac{1}{r} \pd{\psi}{\theta}, 
\end{equation}

la integramos para obtener $\psi$ como

\begin{equation}
\psi(r,\theta) = \frac{Q}{2\pi} \theta + C(r), 
\end{equation}

donde hemos incluido, como corresponde dado que estamos integrando una ecuación diferencial parcial, una función $C(r)$ de la coordenada radial que resta determinar. Para hacerlo, podemos emplear la condición de armonicidad de $\psi$, que en coordenadas cilíndricas resulta:

\begin{equation}
\frac{1}{r} \pd{}{r} \left(r \pd{\psi}{r} \right) + \frac{1}{r^2}\pd{^{2}\psi}{\theta^2} = 0.
\end{equation} 

Observemos que la dependencia de $\psi$ con $\theta$ es lineal, con lo cuál el segundo sumando en la última expresión resulta nulo. En consecuencia, obtenemos una condición para la función $C(r)$ en la forma

\begin{equation}
\frac{1}{r} \pd{}{r} \left(r \pd{C}{r} \right) = 0.
\end{equation} 

Para que esto suceda debe ser constante el término entre paréntesis, con lo cuál se obtiene fácilmente la función $C(r)$,

\begin{equation}
C(r) = A \log r + B.
\end{equation}

Sin pérdida de generalidad, la constante $B$ puede tomarse igual a cero (y eso haremos), dado que, en la determinación del flujo, sólo  interesan las derivadas de la función corriente. 

Hasta este punto, la función corriente resulta

\begin{equation*}
\psi(r,\theta) = \frac{Q}{2\pi} \theta + A \log r,
\end{equation*}

y las componentes del campo de velocidades son (empleando las condiciones sobre $\psi$)

\begin{align}
u_{r} &= \frac{Q}{2\pi r}, \\
u_{\theta} &= -\frac{A}{r}.
\end{align} 

Todavía resta determinar, sin embargo, el valor de la constante $A$. Para ello, consideremos la circulación del campo de velocidades a lo largo de una circunferencia centrada en el origen, y llamemos $\Gamma$ a tal circulación. Como producto de dicho cálculo obtenemos

\begin{equation}
\Gamma = \oint \: \u \cdot d\vec{l} = 
\oint \: (u_{\theta} \: \ver{\theta}) \cdot d\vec{l} =
\oint \: u_{\theta} \: r \: d\theta = -2\pi A.
\end{equation}

Entonces la función corriente resulta, finalmente:

\begin{equation}
\psi(r,\theta) = \frac{Q}{2\pi} \theta - \frac{\Gamma}{2 \pi} \log r.
\end{equation}

## Análisis y discusión de los resultados

### Potencial complejo y campo de velocidades

El potencial complejo, $w(z) = \phi + i \psi$ se obtiene a partir de la última expresión obtenida en la sección precedente. El cálculo arroja

\begin{equation}
w(z) = \frac{Q-i\Gamma}{2\pi} \log z.   
\end{equation}

Cabe destacar que este potencial describe un campo de velocidades asociado a la presencia *en el origen de coordenadas* de una fuente de caudal
constante $Q$ y de un vórtice de circulación $\Gamma$. Extender este resultado al caso en que dicho par se encuentre ubicado en otro punto arbitrario del espacio, $z_0$, resulta trivial:

\begin{equation}
w(z) = \frac{Q-i\Gamma}{2\pi} \log (z-z_0).   
\end{equation}

aunque innecesario para los propósitos de este ejercicio en el que nos
proponemos analizar el flujo asociado unicamente a dichas singularidades.

Luego, el campo de velocidades asociado resulta,

\begin{equation}
\u = u_{r} \: \ver{r} + u_{\theta} \: \ver{\theta} = \frac{Q}{2\pi r} \: \ver{r} 
+ \frac{\Gamma}{2\pi r} \: \ver{\theta}.
\end{equation} 

Es inmediato reexpresar este campo de velocidades en coordenadas cartesianas, de lo
que se obtiene:

\begin{align}
u_{x} &= \frac{1}{2\pi \sqrt{x^{2}+y^{2}}} \left( Q \: x - \Gamma \: y \right), \\
u_{y} &= \frac{1}{2\pi \sqrt{x^{2}+y^{2}}} \left( \Gamma \: x + Q \: y \right).
\end{align}

El campo de velocidades para un juego de valores particular $Q$ y $\Gamma$ se ilustra en la Figura 1. Siendo en este caso $Q=2 > \Gamma =1$, circulación y caudal neto son comparables en magnitud, y los efectos de ambos se observan claramente en la figura. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Q = 2
G = 1

xv   = np.linspace(-1,1,14)
yv   = np.linspace(-1,1,14)
x, y = np.meshgrid(xv,yv)

ux   = 1/2/np.pi/np.sqrt(x**2+y**2)*(Q*x-G*y)
uy   = 1/2/np.pi/np.sqrt(x**2+y**2)*(G*x+Q*y)

fig = plt.figure(figsize=(5,5))
plt.quiver(x,y,ux,uy)
plt.plot(0,0,'ro')
plt.grid('on')
plt.gca().set_aspect('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Figura 1. Mapa del campo de velocidades calculado.** El punto rojo en el origen de coordenadas denota la posicion de la singularidad que genera el flujo.

### Líneas de corriente

En el marco del presente formalismo resulta muy sencillo calcular las líneas de corriente, dado que se corresponden con curvas sobre las cuales la función de corriente tiene un valor constante. Esto surge de considerar la variaci\'on de $\psi$ a lo largo de una línea de corriente,

\begin{equation}
\grad \psi \cdot \u = u_{x} \pd{\psi}{x} + u_{y} \pd{\psi}{y} = 
\pd{\psi}{y} \pd{\psi}{x} - \pd{\psi}{x} \pd{\psi}{y} = 0,
\end{equation}

donde la última igualdad surge de la definición de $\psi$ en términos de $\u$. La función de corriente $\psi$ es entonces constante sobre las líneas de corriente, por lo que
la expresión

\begin{equation}
\psi(\vec{x}) = K, 
\end{equation}

siendo $K$ una constante, es la ecuación de las líneas de corriente.

En el caso particular que estamos analizando, las líneas de corriente del flujo son aquellas curvas en el plano dadas por

\begin{equation}
r(\theta) = \exp \left[ \frac{2\pi}{\Gamma} \left( \frac{Q}{2\pi} \theta - K \right) \right], 
\end{equation}

para un valor de $K$ dado. Esto significa que la ecuación que obtuvimos representa en realidad una *familia de curvas* (las líneas de corriente) parametrizada por el valor de la función corriente sobre cada una de ellas, valor que hemos llamado $K$. 
En virtud de la estacionariedad del flujo, sabemos además que estas curvas 
coinciden con las trayectorias de las partículas de fluido y las líneas
de trazas. 

Geométricamente, la ultima ecuacion nos indica que las líneas de corriente *son espirales logarítmicas* (cuya forma es idéntica, e.g., a la de una sección de un caracol); el panel izquierdo de la Figura 2 (bajo este párrafo) muestra algunas de dichas líneas para valores particulares de $Q, \: \Gamma$ y $K_i$. Complementariamente, el panel derecho nos muestra, con una imagen satelital, un ejemplo de este tipo de espiral observada en la atmósfera terrestre.

```python
# Figura.
fig = plt.figure(figsize=(12,5))

# Panel izquierdo de la figura.
# Tres lineas de corriente.

# Definimos los valores de los parametros.
G  = -4     # valor de la circulacion 
Q  = 0.8    # valor del caudal
K1 = 3.7    # valor para la 1ra linea de corriente considerada
K2 = 3.5    # valor para la 2da linea de corriente considerada
K3 = 3.9    # valor para la 3ra linea de corriente considerada 

# Creamos la grilla sobre la que graficaremos.
theta = np.linspace(-20*np.pi, 20*np.pi, 1000)

# Calculamos las 3 lineas de corriente consideradas.
r1 = np.exp( (2*np.pi/G)*(Q/2/np.pi*theta) - K1)
r2 = np.exp( (2*np.pi/G)*(Q/2/np.pi*theta) - K2)
r3 = np.exp( (2*np.pi/G)*(Q/2/np.pi*theta) - K3)

# Graficamos la figura del panel izquierdo.
ax1 = plt.subplot(121)
plt.plot(r1*np.cos(theta), r1*np.sin(theta), 'g', linewidth=2)
plt.plot(r2*np.cos(theta), r2*np.sin(theta), 'b', linewidth=2)
plt.plot(r3*np.cos(theta), r3*np.sin(theta), 'r', linewidth=2)

plt.xlim(-10,10)
plt.ylim(-10,10)
ax1.set_aspect('equal')
plt.grid(True)
plt.xlabel('x [u.a.]')
plt.ylabel('y [u.a.]')

# Panel derecho de la figura.
# Imagen del MODIS, para comparacion cualitativa.
import matplotlib.image as mpimg
imagen = mpimg.imread("Figuras/Potencial_Complejo/Hadis.jpg")
ax2 = plt.subplot(122)
ax2.set_aspect('equal')
plt.imshow(imagen)
plt.xlabel('x [px]')
plt.ylabel('y [px]')

# Mostramos ambos paneles de la figura.
plt.show()
```

**Figura 2. Líneas de corriente en forma de espirales logarítmicas.**

**Panel Izquierdo**: Ejemplo de líneas de corriente del flujo para el caso $Q = 0.8, \Gamma = -4$. Se muestran tres líneas de corriente con valores $K$ dados por: $K_1 = 3.5$ (en azul), $K_2 = 3.7$ (en verde) y $K_3 = 3.9$ (en rojo).

**Panel Derecho**: Imagen satelital que muestra una depresión centrada en el estrecho de Dinamarca entre Islandia y Groenlandia. Los vientos, atraídos por la zona de baja presión en el centro, se curvan por el efecto de la fuerza de Coriolis, y observamos la espiral con la apariencia de caracol desde el espacio. La imagen fue tomada por el instrumento Aqua MODIS (NASA), el 4 de septiembre de 2003. Para visualizar una versión de alta resolución de esta imagen, visitar [este link](http://visibleearth.nasa.gov/view_rec.php?id=6204).



**En el caso general, vemos que el requerimiento de que una fuente puntual emita un caudal $Q$ constante nos arroja un flujo que -al menos en principio- podría presentar una circulación no nula. Es decir: obtuvimos como resultado de nuestro análisis
que una circulación no nula es compatible con un flujo originado por una fuente puntual cuyo caudal es constante.**

Los dos casos límite que surgen de nuestros resultados son los siguientes:

- $\Gamma = 0$, el flujo resultante es el correspondiente a una **fuente isótropa** de caudal $Q$ constante que emite en forma radial. En ese caso la función corriente y el potencial complejo resultan

\begin{align}
\psi_{\text{fuente}} &= \frac{Q}{2\pi} \theta, \\
w_{\text{fuente}}(z) &= \frac{Q}{2\pi} \log z.
\end{align}

- $Q = 0$, caso en el cuál el flujo corresponde al de un vórtice libre e isótropo de circulación $\Gamma$. La función corriente y el potencial complejo resultan en este caso

\begin{align}
\psi_{\text{vórtice}} &= - \frac{\Gamma}{2\pi} \log r, \\
w_{\text{vórtice}}(z) &= -i \frac{\Gamma}{2\pi} \log z.
\end{align}

## Sumario

En este problema pudimos calcular la función de corriente, las líneas de corriente (y graficarlas en Python) así como también obtener una expresión para el potencial complejo de una fuente de caudal constante $Q$. 

Este caso es de interés por
dos razones. Por un lado, el ejemplo sirve como ilustración del método general para
el cálculo del potencial complejo de un flujo *singular*. Por el otro, vemos que
hemos calculado, en un solo ejemplo, el potencial complejo para los dos **ladrillos fundamentales** de los que estarán constituídos todos los flujos que consideraremos en esta práctica: una fuente isótropa y un 
vórtice (los dos casos límite descriptos en la sección precedente). Cualquier flujo que resulte combinación de ellos (p.ej., dipolos, pero también cualquier otra 
distribución de fuentes, sumideros y/o vórtices), podrá calcularse 
fácilmente a partir del resultado que obtuvimos, dado que las ecuaciones para 
$\phi$ y $\psi$ responden al principio de superposición.
