## Introduccion

### Problema I: Modelo de Ciclon

Considere el campo de velocidades de un fluido consistente en un nucleo cilindrico muy alargado de base circular de radio $a$, que rota rigidamente sobre su eje principal con velocidad angular constante $\Omega$. Fuera del nucleo, el campo de velocidades es tambien azimutal, pero irrotacional; siendo continuo en $r=a$. En estas condiciones:
- Determine los campos de velocidad y de vorticidad para todo valor de $r$.
- Determine la distribucion espacial del campo de presion para todo $r$, en funcion de la presion muy lejos del eje (que denominaremos $p_\infty$). Encuentre que condicion debe satisfacer $\Omega$ respecto del valor de la presion $p_\infty$.

### Resolucion

#### Determinacion del campo de velocidad azimutal

Comencemos por construir el campo de velocidad azimutal asociado al flujo descripto por el enunciado. Dado que el flujo presenta simetria cilindrica, usaremos ese sistema de coordenadas para describirlo matematicamente. En esta descripcion y, segun nos dice el enunciado, la velocidad azimutal en el interior (que llamaremos $u_\theta^{int}$) debe corresponder a la de una rotacion rigida con velocidad angular $\Omega$, es decir:

$$ u_\theta^{int}(r) = r \: \Omega \: \hat{r}.$$

Por otro lado, en el exterior la velocidad es tambien azimutal,

$$
\vec{u}^{ext} = u_\theta^{ext}(r) \: \hat{\theta}, \\
$$
pero irrotacional, es decir:
$$\omega_z^{ext} = \vec{\nabla} \times \big( u_\theta^{ext} \: \hat{\theta} \big) = 0.$$

Notemos que esta segunda igualdad es una ecuacion diferencial (de primer orden) en derivadas parciales para $u_\theta^{ext}$. La expresion en componentes para el rotor de un campo vectorial $\vec{A}$ en coordenadas cilindricas viene dada por:

$$
\bigg( \vec{\nabla} \times \vec{A}\bigg)_{cil} = 
\bigg( \frac{1}{r} \frac{\partial A_z}{\partial \theta} - \frac{\partial A_\theta}{\partial z} \bigg) \: {\hat{r}} + 
\bigg( \frac{\partial A_r}{\partial z} - \frac{\partial A_z}{\partial r} \bigg) \: {\hat{\theta}} + 
\frac{1}{r} \bigg[ \frac{\partial \big(r A_\theta \big)}{\partial r} - \frac{\partial A_r}{\partial \theta}
  \bigg] \: {\hat{z}}.
$$

En el caso que nos ocupa, el campo vectorial de velocidades $u_\theta^{ext}$ solo tiene componente en la direccion $\hat{\theta}$, y depende exclusivamente de la coordenada radial $r$, luego:

$$
\vec{\nabla} \times \big( u_\theta^{ext} \: \hat{\theta} \big)  = 
\frac{1}{r} \bigg[ \frac{\partial \big(r \: u_\theta^{ext} \big)}{\partial r} \bigg] \: {\hat{z}}.
$$

Ahora bien, dado la vorticidad en el exterior del nucleo del vortice debe ser nula en toda posicion del espacio, resulta

$$
\frac{\partial \big(r \: u_\theta^{ext} \big)}{\partial r} = 0,
$$

es decir que el monomio entre parentesis debe ser igual a una constante $C$, luego el campo de velocidades en la region exterior del vortice es

$$
u_\theta^{ext} = \frac{C}{r}.
$$

Observemos que el valor de la constante $C$ se fija imponiendo la continuidad del campo de velocidades en $r=a$:

\begin{align}
u_\theta^{ext}(r=a) &= u_\theta^{int}(r=a) \Rightarrow \\
\frac{C}{a} &= a \: \Omega, \\
 \Rightarrow C &= \Omega \: a^2. 
\end{align}

Finalmente, el campo de velocidades buscado es   

$$
\vec{u} = u_\theta(r) \: \hat{\theta} = \begin{cases}
r \: \Omega, & r < a, \\
\frac{\Omega a^2}{r}, & r > a,
\end{cases}
$$

Este campo de velocidad corresponde al denominado **Vortice de Rankine**, que resulta el modelo mas sencillo de un vortice real. Los vortices reales estan caracterizados por nucleos ('cores') de extension espacial relativamente pequena, en los cuales se concentra la vorticidad, mientras que en el exterior el flujo es -en buena aproximacion- irrotacional. El *core* no es, en general, exactamente circular, ni tampoco la vorticidad en el interior es homogenea. En estos dos aspectos, el vortice de Rankine es solo un *modelo idealizado de un vortice real*. 

En ultimo lugar, el campo de vorticidad asociado al vortice de Rankine resulta dado por

$$
\vec{\omega} = \omega_z(r) \: \hat{z} = \begin{cases}
2 \: \Omega, & r < a, \\
0, & r > a.
\end{cases}
$$

Las siguientes lineas de codigo Python sirven el proposito de graficar los resultados obtenidos para los campos de velocidad azimutal y de vorticidad del vortice de Rankine. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

a = 1
W = 1
rint = np.linspace(0, a, 100)
rext = np.linspace(a, 5*a, 900)

fig = plt.figure(figsize=(14,4))
ax1 = fig.add_subplot(121)
ax1.plot(rint, W*rint, 'r', linewidth=3, label='interior del vortice')
ax1.plot(rext, W*a**2/rext, 'b', linewidth=3, label='exterior del vortice')
plt.xlabel(r'$r/a$', fontsize=20)
plt.ylabel(r'$\frac{u_{\theta}(r/a)}{\Omega a}$', fontsize=24)
plt.ylim([0, W*a*1.2])
plt.grid('on')
plt.title(r'Perfil radial de velocidad azimutal')
plt.legend()

ax2 = fig.add_subplot(122)
ax2.plot(rint, 2*W*np.ones(rint.shape)/2/W, 'r-', linewidth=3, label='interior del vortice')
ax2.plot(rext, np.zeros(rext.shape), 'b-', linewidth=3, label='exterior del vortice')
plt.xlabel(r'$r/a$', fontsize=20)
plt.ylabel(r'$\frac{\omega_z(r/a)}{2\Omega}$', fontsize=24)
plt.ylim([0, 1.2])
plt.grid('on')
plt.title(r'Perfil radial de la componente vertical de la vorticidad')
plt.legend()

plt.show()
```

FIG. 1. **Campos de velocidad azimutal y de vorticidad vertical para el vortice de Rankine.** El panel de la izquierda muestra el campo de velocidad azimutal, que crece linealmente en el interior (nucleo) del vortice hasta alcanzar un valor de $(\Omega a)$ en el borde. En el exterior, la velocidad azimutal varia de forma inversamente proporcional a la distancia al eje del vortice. El panel de la derecha ilustra el campo de vorticidad asociado. En el nucleo del vortice, la vorticidad es constante e igual a $(2\Omega)$, lo que corresponde a una rotacion rigida de velocidad angular $\Omega$. En el exterior, el flujo es irrotacional. 

*En esta figura, y para mayor generalidad, tanto el campo de velocidad como el de vorticidad se grafican en forma adimensional y dependiendo del cociente $r/a$.*

#### Determinacion del campo de presiones

Comenzamos planteando la ecuacion general para el flujo incompresible de un fluido ideal (sin viscosidad), es decir:

$$
\frac{d\vec{u}}{dt} = - \frac{1}{\rho} \vec{\nabla}p + \vec{g},
$$

donde la derivada material puede separarse en sus dos componentes

$$
\frac{\partial\vec{u}}{\partial t} + \big( \vec{u}\cdot \vec{\nabla} \big) \vec{u} = - \frac{1}{\rho} \vec{\nabla}p + \vec{g}.
$$

Dado que estamos considerando un flujo **estacionario**, el termino de aceleracion temporal de la derivada material (la derivada parcial temporal) resulta identicamente nula, con lo cual la ecuacion para el balance hidrostatico del ciclon resulta:

$$
\big( \vec{u}\cdot \vec{\nabla} \big) \vec{u} = - \frac{1}{\rho} \vec{\nabla}p + \vec{g}.
$$

En el sistema de coordenadas cilindricas en el que estamos trabajando, el operador diferencial $(\vec{u}\cdot \vec{\nabla})$ tiene una expresion dada por

$$
\big( \vec{u}\cdot \vec{\nabla} \big)_{cil} = \frac{u_\theta}{r} \frac{\partial}{\partial \theta}, 
$$

luego el lado izquierdo de la ecuacion de balance hidrostatico se escribe, en nuestro caso, como

$$
\big( \vec{u}\cdot \vec{\nabla} \big) \vec{u} = \frac{u_\theta}{r} \frac{\partial}{\partial \theta} \bigg[ u_\theta(r) \: \hat{\theta} \bigg] = \frac{u_\theta}{r} \bigg[ \frac{\partial u_\theta}{\partial \theta} \: \hat{\theta}  + u_\theta(r) \: \frac{\partial \hat{\theta}}{\partial \theta}  \bigg].
$$

Consideremos entonces los sumandos entre corchetes en la ultima expresion. En primer lugar, sabemos que la componente azimutal del campo de velocidad solo depende de $r$, con lo cual el primer sumando es nulo. Por otro lado, sabemos que 

$$
\frac{\partial \hat{\theta}}{\partial \theta} = -\hat{r},
$$

de lo que resulta

$$
- \frac{u_\theta(r)}{r} \: \hat{r}  = - \frac{1}{\rho} \vec{\nabla}p - g \: \hat{z}.
$$

Esta ecuacion vectorial puede descomponerse en tres ecuaciones escalares para el campo de presiones:

\begin{align}
\frac{\partial p}{\partial r} &= \rho \frac{u_\theta^2}{r}, \\
\frac{\partial p}{\partial \theta} &= 0, \\
\frac{\partial p}{\partial z} &= -\rho g.
\end{align}

De la segunda de estas tres ecuaciones deducimos que la presion no puede ser funcion de la coordenada azimutal, solo de la coordenada radial y de la vertical, es decir $p \equiv p(r,z)$.

La solucion a la primera ecuacion, por otro lado, depende de la region del espacio en la que estemos interesados, dado que incorpora el campo de velocidades azimutal que presenta dos  comportamientos diferenciados en el interior y en el exterior del nucleo del vortice. 

La tercera ecuacion, por el contrario, no depende de la region del espacio (es valida *con los mismos coeficientes* tanto en el interior como en el exterior del 'core' del vortice), y señala que la dependencia de la presion con la coordenada vertical corresponde a un perfil lineal cuya magnitud local decrece conforme nos alejamos del suelo (que podemos fijar arbitrariamente como el plano para el cual $z=0$). 

Integremos entonces la primera ecuacion para la presion en la region correspondiente al nucleo del vortice. Alli resulta:

$$\frac{\partial p}{\partial r} = \rho \Omega^2 r \quad \Rightarrow \quad p(r,z) = \frac{1}{2} \rho \Omega^2 r^2 + F(z) + C,$$

donde $F(z)$ es una funcion arbitraria de la variable $z$ sin termino constante, y $C$ es una constante a fijar *a posteriori*. Si ahora integramos la tercera ecuacion para la presion, obtenemos

$$p(r,z) = -\rho g z + G(r) + D, $$

siendo $G(r)$ una funcion arbitraria de $r$ sin termino constante y $D$ una constante a determinar. Reuniendo estos ultimos dos resultados en una misma expresion, obtenemos

$$ p_{int}(r,z) = \frac{1}{2} \rho \Omega^2 r^2 - \rho g z + C_1, \quad r < a. $$

Repitiendo este procedimiento de integracion para la region externa al nucleo del vortice, resulta facil obtener la siguiente expresion para el campo de presiones en dicha zona:

$$p_{ext}(r,z) = - \frac{1}{2} \frac{\rho \Omega^2 a^4}{r^2}  - \rho g z + C_2, \quad \quad r>a. $$

**Antes de continuar** (p.ej., imponiendo la continuidad del campo de presiones entre el interior y el exterior del nucleo del vortice), **reflexionemos acerca del sentido fisico de las constantes** $C_1$ y $C_2$. Por un lado, observemos que $C_1$ resulta ser igual a la presion en el interior del nucleo cuando hacemos $r=0$ y $z=0$, es decir:

$$p_{int}(r=0, z=0) = C_1,$$

con lo cual la constante $C_1$ corresponde entonces a la presion que se mediria en el centro de un tornado ($r=0$) a la altura (altitud) del suelo, $z=0$. Llamemos a dicha presion $p_0$, definida como $p_0 \equiv p_{int}(0,0)$.

Similarmente, observemos que la constante $C_2$ corresponde a la presion a la altura del suelo pero a distancias infinitamente alejadas el centro del vortice. Es decir,

$$p_{ext}(r\rightarrow \infty, z=0) = C_2.$$

Con este resultado, podemos denotar entonces a $C_2$ con el simbolo $p_\infty$, definido de forma tal que $p_\infty \equiv p_{ext}(r\rightarrow \infty, z=0)$.

En funcion de estas ideas, la expresion para el campo de presiones en todo el espacio resulta:

$$
p(r,z) = \begin{cases}
\frac{1}{2} \rho \Omega^2 r^2 - \rho g z + p_0, & r < a, \\
- \frac{1}{2} \frac{\rho \Omega^2 a^4}{r^2}  - \rho g z + p_\infty, & r > a.
\end{cases}
$$

Ahora bien, el campo de presiones debe ser continuo entre el interior y el exterior del nucleo del vortice, con lo cual:

$$p_{int}(r=a, z) = p_{ext}(r=a,z),$$

que implica

$$\frac{1}{2} \rho \Omega^2 a^2 - \rho g z + p_0 = - \frac{1}{2} \frac{\rho \Omega^2 a^4}{a^2}  - \rho g z + p_\infty,$$

a partir de lo cual es posible obtener una expresion que relaciona las presiones $p_0$ y $p_\infty$:

$$ p_0 = p_\infty - \rho \Omega^2 a^2.$$

Notemos que este resultado muestra que la presion en $r=0$ es menor a aquella en $r=\infty$, en una cantidad dada por el producto $(\rho \: \Omega^2 a^2)$, lo que explica la muy baja presion que se observa en el nucleo de un tornado. 

En las proximas lineas de codigo Python graficamos el resultado que acabamos de obtener para el campo de presiones, considerando valores arbitrarios de las constantes. En particular, utilizaremos
$\rho$=1, $\Omega$=10, $g$=9.8 y $a$=1. Asimismo, debemos fijar el valor de $p_0$ o bien de $p_\infty$; en este caso fijamos, arbitrariamente, $p_\infty$=1000. 

Dado que $p(r,z)$ es una funcion de dos variables, vamos a visualizarla en un mapa 2D del plano $r$-$z$ en el que los valores de la funcion estaran codificados localmente (en cada punto de dicho plano) mediante un color. En particular, la paleta de colores que usaremos va desde el azul (identificando valores bajos de $p$) hasta el rojo (que denota valores elevados del mismo campo).

```python
# Definimos los valores numericos del vortice.
rho = 1
W = 10
a = 1
g = 9.8

# Definimos las condiciones de contorno.
pinf = 1000
p0 = pinf - rho*W**2*a**2

# Definimos la grilla espacial de trabajo.
Npuntos = 20
riv = np.linspace(0, a, Npuntos)
ziv = np.linspace(0, 10*a, 10*Npuntos)
rev = np.linspace(a, 5*a, 4*Npuntos)
zev = np.linspace(0, 10*a, 10*Npuntos)
ri, zi = np.meshgrid(riv, ziv)
re, ze = np.meshgrid(rev, zev)

# Calculamos la presion interna y externa.
pint = 1/2*rho*W**2*ri**2 - rho*g*zi + p0
pext = -1/2*rho*W**2*a**4/re**2 - rho*g*ze + pinf

# Valores maximos y minimos de la presion, 
#  utiles para la barra de colores del grafico.
pmax = np.amax(pext)
pmin = np.amin(pint)
```

```python
# Construimos el grafico de presion p(r,z).
plt.figure(figsize=(8,5))
plt.pcolor(ri,zi,pint, vmin=pmin, vmax=pmax)
plt.pcolor(re,ze,pext, vmin=pmin, vmax=pmax)
plt.colorbar()
ci = plt.contour(ri,zi,pint, colors='k')
ce = plt.contour(re,ze,pext, colors='k')
plt.clabel(ci, inline=1, fontsize=10)
plt.clabel(ce, inline=1, fontsize=10)
plt.xlabel(r'$r$', fontsize=20)
plt.ylabel(r'$z$', fontsize=20)
plt.title('$p(r,z)$ (colormap) e isobaras (lineas)', fontsize=20)
plt.grid('on')
plt.show()
```

FIG. 2. **Campo de presiones** $p(r,z)$ ** (en color) e isobaras (lineas negras) para el vortice de Rankine.** Observemos que la presion disminuye conforme nos acercamos al eje de simetria del vortice, y tambien cuando nos movemos alejandonos del suelo $z$=0.

El grafico muestra que la presion del fluido disminuye conforme nos acercamos al eje de simetria del vortice $r=0$, y tambien cuando nos movemos alejandonos del suelo (plano con $z=0$). A partir de este mapa de la distribucion de presiones es posible entonces comprender el efecto de 'succion' que un tornado tiene sobre las particulas y/o los objetos que encuentra a su paso. Para hacer mas claro este hecho, las lineas de codigo Python que siguen calculan el gradiente de presiones $\vec{\nabla}p \equiv (\partial_r p, \partial_z p)$ y grafican el campo de presiones (en colormap) pero ahora en conjuncion con el campo vectorial -$\vec{\nabla}p$. 

```python
# Calculamos el gradiente de presiones.
pintz, pintr = np.gradient(pint, ziv[1]-ziv[0], riv[1]-riv[0])
pextz, pextr = np.gradient(pext, zev[1]-zev[0], rev[1]-rev[0])
# Graficamos.
plt.figure(figsize=(8,5))
plt.pcolor(ri,zi,pint, vmin=pmin, vmax=pmax)
plt.pcolor(re,ze,pext, vmin=pmin, vmax=pmax)
plt.colorbar()
plt.quiver(riv[::5],ziv[0::20],-pintr[0::20,::5],-pintz[0::20,::5], pivot='mid')
plt.quiver(rev[::5],zev[0::20],-pextr[0::20,::5],-pextz[0::20,::5], pivot='mid')
plt.xlabel(r'$r$', fontsize=20)
plt.ylabel(r'$z$', fontsize=20)
plt.title(r'$p(r,z)$ (colormap) y $-\vec{\nabla}p$ (flechas)', fontsize=20)
plt.grid('on')
plt.show()
```

FIG. 3. **Campo de presiones** $p(r,z)$ ** (en color) y $-\vec{\nabla}p$ (flechas negras) para el vortice de Rankine.**

Esta ultima figura refuerza entonces la idea que expusimos en el parrafo anterior acerca del efecto de un tornado (que un vortice de Rankine modela) sobre su entorno. Dado que el nucleo del vortice es un centro de baja presion (relativa), el gradiente de presion en el medio apunta hacia el, conviertiendolo en un centro de succion. 

## Problema II: Vortice de vaciado

Repensemos ahora el Problema I propuesto inicialmente, pero como un problema en el cual, en lugar de modelar un ciclón, buscamos modelar el campo de velocidades asociado al vaciado de un líquido contenido en un recipiente (abierto a la atmósfera) a través de un agujero en su base. Atención: hacemos esto pensando en que el modelo que construimos es válido **lejos de la abertura de salida**. No nos ocuparemos, al menos en este momento, de lo que pasa cerca de la abertura de salida (lo haremos más adelante en el curso).

### Forma de la superficie libre

Les propuse entonces –como ejercicio optativo– calcular, a partir de las expresiones obtenidas en clase para el campo de velocidades y la presión, la forma de la superficie libre que se encuentra a presión atmosférica $p_0$.

En primer lugar, calculemos la diferencia en altura de la superficie libre en el origen y muy lejos de él:

\begin{align}
r = 0: \quad p_0 &= - \rho g z_0 + C_1 \\
r \rightarrow \infty : \quad p_0 &= - \rho g z_\infty + C_2    
\end{align}

de lo cuál deducimos que

$$ z_\infty -z_0 = \frac{C_2-C_1}{\rho g} = \frac{\Omega^2 a^2}{g}$$

Para determinar la forma de la superficie libre, utilizamos que sobre dicha superficie la presión es la presión atmosférica. Entonces en cada región ($r < a$ y $r > a$), calculamos la altura a la que la superficie libre se encuentra a partir de la expresión para la presión:

\begin{align}
p_0 &= \frac{\rho \Omega^2 r^2}{2} - \rho g z + C_1, \quad \quad r < a, \\
p_0 &= -\frac{\rho \Omega^2 a^2}{2\: r^2} - \rho g z + C_2, \quad \quad r>a. 
\end{align}

De lo que se obtiene la posición vertical de la superficie libre, $z_{sl}(r)$, en cada región:

$$
z_{sl}(r) =
\begin{cases} 
\frac{\Omega^2 r^2}{2g} + \frac{C_1-p_0}{\rho g} & r < a \\
\frac{-\rho \Omega^2 a^4}{2gr^2} + \frac{C_2-p_0}{\rho g} & r > a 
\end{cases}
$$
 

De esta última expresión, y dado que $C_1$ y $C_2$ son constantes, ya podemos observar la dependencia funcional de $z_{sl}(r)$ en cada región. En el interior del vórtice (en su núcleo o core), tenemos $z_{sl}(r) \propto r^2$, mientras que en la región exterior es $z_{sl}(r) \propto r^{−2}$. Notemos entonces, como les anticipé en clase, que obtuvimos un perfil para la superficie libre cuya curvatura cambia conforme estemos dentro o fuera del núcleo del vórtice.

Para obtener una expresión completa en donde sólo aparezcan explícitamente las variables del problema (y no, p.ej., $C_1$ y $C_2$), establezcamos el origen de alturas de forma tal que coincida con la posicion vertical de la superficie libre en $r = 0$. Esto es, hagamos: $z_{sl}(r = 0) = 0$, lo cuál es siempre lícito dado que corresponde a fijar nuestro origen de coordenadas. Luego es fácil ver que $C_1 = p_0$, y finalmente con un poco de álgebra elemental obtenemos:

$$z_{sl}(r)=
\begin{cases}
\frac{\Omega^2 r^2}{2g}, & r < a, \\
\frac{\Omega^2 a^2}{g} \bigg( 1 - \frac{a^2}{2 \: r^2 } \bigg), & r > a,
\end{cases}
$$

que nos da el perfil de la superficie libre buscado. La Fig. 1 muestra gráficamente la dependencia de este perfil con la coordenada radial, para un juego de valores particulares de $a$, $Ω$ y $g$.

```python
a = 2
W = 1
g = 9.8

rint = np.linspace(0, a, 100)
rext = np.linspace(a, 5*a, 900)

zsl_int = W**2*rint**2/2/g
zsl_ext = W**2*a**2/g*(1 - a**2/2/rext**2)

plt.figure()
plt.plot(rint, zsl_int, 'r', linewidth=2)
plt.plot(rext, zsl_ext, 'b', linewidth=2)
plt.xlabel(r'$r$', fontsize=20)
plt.ylabel(r'$z_{sl}(r)$', fontsize=20)
plt.axis([0, 5*a, 0, 1.25*zsl_ext[-1]])
plt.title('Perfil radial de la superficie libre')
plt.grid('on')
plt.show()
```

FIG. 4. **Perfil de la superficie libre en función de la coordenada radial**, correspondiente a $Ω =$ 1, $a =$ 2, $g =$ 9.8. Notar la diferencia entre escalas en los ejes horizontal y vertical.

Observamos entonces que, en el núcleo, el perfil de la superficie libre es cóncavo y parabólico (la superficie de un líquido en rotación rígida adopta el mismo perfil; recuerden el Problema del 'taquímetro hidrostático'). **En la región externa**, por el contrario, **la superficie libre es convexa dado que la velocidad disminuye con la distancia**: el perfil en esa zona resulta de tipo hiperbólico.

## Acerca del modelo de vortice de Rankine

### Comparacion con tornados

Un gran numero de estudios observacionales *in situ* han mostrado que el modelo simple de Rankine resulta una buena aproximacion al campo de velocidad (del viento local) asociado con un tornado. 

En esta seccion deseo ilustrar con un ejemplo el grado de acuerdo entre el modelo teorico de Rankine y datos experimentales reportados recientemente acerca de un tornado inusualmente intenso que causo daños de grado F4 en la [escala Fujita](http://en.wikipedia.org/wiki/Fujita_scale) (acabando con la vida de varias personas) en Mulhall, Estados Unidos, el 3 de mayo de 1999. El tornado Mulhall (como dio en llamarsele) es uno de los tornados del ['Oklahoma tornado outbreak']( 'http://en.wikipedia.org/wiki/1999_Oklahoma_tornado_outbreak), y forma parte de los [*tornados record*](http://en.wikipedia.org/wiki/Tornado_records) en la historia de Estados Unidos. 

El tornado Mulhall fue interceptado por el radar movil DOW ('Doppler on Wheels') y seguido durante 9 kilometros, resultando en datos de alta resolucion volumetrica con una cadencia temporal de 55 segundos, hasta una altitud de 1.5 kilometros, durante un periodo de 14 minutos. Estas mediciones permitieron, por primera vez, determinar la estructura y evolucion tridimensionales de un tornado empleando la tecnica GBVTD (ground-based velocity track display). Por medio de esta tecnica experimental, pudo determinarse la circulacion del vortice y la velocidad azimutal del viento para cada radio y altura dentro del rango de medicion del instrumento. Asimismo, pudieron obtenerse estimaciones para el momento angular del tornado, su vorticidad y el campo de presiones asociado. 

**Debido a la alta calidad de las mediciones de las que se dispone para el tornado Mulhall, el mismo resulta ideal para ilustrar a que punto el modelo de Rankine (y las magnitudes derivadas a partir de el) se corresponden con lo observado experimentalmente.**

Brevemente, las caracteristicas del tornado Mulhall son las siguientes. La circulacion primaria del tornado consistio de una componente azimutal de vientos de velocidad pico de 84 m/s (~300 km/h), con un nucleo de dimensiones entre 500 y 1000 metros. En ultimo lugar, cabe mencionar que la diferencia de presion maxima (normalizada a una distancia de 3 kilometros del centro y a 50 metros de altitud) fue de -80 hPa. 

*Un reporte detallado de las caracteristicas del tornado Mulhall puede consultarse en Lee & Wurman (2005).*
